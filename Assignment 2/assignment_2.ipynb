```python
from load_mnist import load_mnist
from utils import NeuralNetwork
import numpy as np
np.random.seed(0)
```

## Load MNIST dataset

```python
X_train, y_train, X_test, y_test = load_mnist()
```

## One-layer model

```python
model_architecture_1_layer = {
        "hidden_layer_1": {"units": 512, "activation_function": "relu"},
    }
neural_network_1_layer = NeuralNetwork(model_architecture_1_layer)
neural_network_1_layer.train_model(
    X_train,
    y_train,
    X_test,
    y_test,
    learning_rate=0.005,
    mini_batch_size=128,
    iterations=100,
)
```

```python
neural_network_1_layer.plot_weights()
```

## Two-layer model

```python
model_architecture_2_layer = {
        "hidden_layer_1": {"units": 256, "activation_function": "relu"},
        "hidden_layer_2": {"units": 128, "activation_function": "relu"},
    }
neural_network_2_layer = NeuralNetwork(model_architecture_2_layer)
neural_network_2_layer.train_model(
    X_train,
    y_train,
    X_test,
    y_test,
    learning_rate=0.005,
    mini_batch_size=128,
    iterations=100,
)
```

```python
neural_network_2_layer.plot_weights()
```
